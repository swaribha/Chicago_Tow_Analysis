```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
vehicals_df_clean=pd.read_csv("data/towed_vehical_clean.csv")
vehicals_df_clean
```

```python
abandon_vehicals=pd.read_csv("Data/311_Abandoned_Vehicles.csv")
abandon_vehicals.tail(30)
```

```python
#extracting the data for last 6 months
abandon_vehicals['Creation Date']=pd.to_datetime(abandon_vehicals['Creation Date'], infer_datetime_format=True)

date1 = datetime.date(2018,10,21)


# df = df.loc[(df['date']>date1) & (df['date']<date2)]
abandon_vehicals=abandon_vehicals.loc[(abandon_vehicals['Creation Date'] >date1)]
print(abandon_vehicals.shape)
abandon_vehicals.head(100)
```

```python
# dropping the duplicates request for same vehicle
abandon_vehicals=abandon_vehicals.loc[~abandon_vehicals['Status'].str.contains("Dup")]
print(abandon_vehicals.shape)
```

```python
abandon_vehicals.rename(columns={"License Plate":"Plate"},inplace=True)
```

```python
abandon_vehicals=abandon_vehicals[['Creation Date', 'Plate']]
abandon_vehicals.dropna(subset=['Plate'],inplace=True)
abandon_vehicals
```

```python
# dropping the data where Lincense plate is not known 
abandon_vehicals=abandon_vehicals.loc[~abandon_vehicals['Plate'].str.contains("DON'T KNOW|NONE|NLP|UNKNOWN|NO|DON'T HAVE|XXXXXX|UNK|0|\
                                                                              UNKOWN NUMBER|MIssing|Plates|\\.|\\?",case=False)]
#dropping the duplicates plate data as we are concerned with unique plates
abandon_vehicals.sort_values('Creation Date',inplace=True)
abandon_vehicals.drop_duplicates('Plate',inplace=True)
abandon_vehicals
```

```python
Abandon_towed_days=pd.merge(vehicals_df_clean,abandon_vehicals ,on='Plate',how='inner')
Abandon_towed_days.drop(columns="Unnamed: 0",inplace=True)
Abandon_towed_days
```

```python
# Calculate the days from date reported to date dowed
Abandon_towed_days['Tow Date']=pd.to_datetime(Abandon_towed_days['Tow Date'], infer_datetime_format=True)
Abandon_towed_days['Days diff']=Abandon_towed_days['Tow Date']-Abandon_towed_days['Creation Date']
# converting time delta object to int days
Abandon_towed_days['Days diff']=Abandon_towed_days['Days diff'].apply(lambda x: x.days)
Abandon_towed_days
```

### Average Days for a car to get Towed from date reported

```python
Average_days=Abandon_towed_days['Days diff'].mean()
Average_days
```

```python
#Visualization of diff between creatin date and tow date
plt.figure(figsize=(15,10))
plt.scatter(Abandon_towed_days["Creation Date"].tolist(),Abandon_towed_days["Tow Date"].tolist(),
            s=Abandon_towed_days['Days diff']+100,color='blue',edgecolor='black',alpha=0.75)
plt.xlabel("Date Reported")
plt.ylabel("Date Towed")
plt.title("Date Reported Vs Date Towed Abandon Car")
plt.savefig("figures/Abandoncars_scatter.png")

```

### Analysis of Percentage of Towed cars Abandoned

```python
tow_abandon=pd.merge(vehicals_df_clean,abandon_vehicals ,on='Plate',how='left')
tow_abandon.drop(columns="Unnamed: 0",inplace=True)
tow_abandon
```

```python
tow_abandon['Abandon flag']=1
tow_abandon.loc[tow_abandon['Creation Date'].isnull(),'Abandon flag']=0

# tow_abandon['test'] = np.where(tow_abandon['Make'].map(len) > tow_abandon['Color'].map(len))#, df['a'].map(len), df['b'].map(len))

sizes=tow_abandon['Abandon flag'].value_counts()
sizes
```

###  Creating the pie chart 

```python
labels=['Non Abandon Cars','Abandon Cars']
colors=['#99ff99', '#66b3ff']
explode=[0,0.3]
plt.pie(sizes, labels=labels, colors=colors, startangle=30,explode=explode,radius=3,autopct="%1.1f%%",shadow=True)
# plt.pie(sizes_gender,colors=colors_gender,radius=0.75,startangle=90)
plt.title("% of Abandon cars towed by Chicago city")
plt.axis('equal')
plt.savefig("figures/Abandoncars_pie.png")
plt.show()
```
